```python
from CNN.kernel import *
from PIL import Image
```

```python
img = Image.open('/Users/julienthomas/Downloads/IMG_1255.jpg')
img.load()
image_data = np.asarray( img, dtype="int32" )
```

```python
filter_array = np.array([[1, 0, -1],
                [0, 0, 0],
                [-1, 0, 1]], dtype=np.int32)
kernel = Kernel((3, 3))
kernel.array = filter_array
f_map = kernel.convolve(image_data)
```

```python
from matplotlib import pyplot as plt
plt.imshow(f_map, interpolation='nearest')
plt.show()
```

```python
from CNN.layer import PoolLayer

pl = PoolLayer((3,3))
pl.inputs.append(image_data)
pl.pool('max')

plt.imshow(pl.output[0], interpolation='nearest')
plt.show()
```

```python
pl = PoolLayer((3,3))
pl.inputs.append(image_data)
pl.pool('average')

plt.imshow(pl.output[0], interpolation='nearest')
plt.show()
```
