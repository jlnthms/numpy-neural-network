```python
from CNN.network import *
from NeuralNetwork.network import *

from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
```

```python
mnist = fetch_openml('mnist_784')
```

```python
image = mnist.data.to_numpy()
reshaped_images = image.reshape(-1, 28, 28)
plt.imshow((reshaped_images[0]), cmap=plt.cm.gray_r, interpolation='nearest')
```

```python
LeNet_layers = [
    ConvLayer(6, (5, 5), 'relu'),
    PoolLayer((2, 2), stride=2, method='average'),
    ConvLayer(16, (5, 5), 'relu'),
    PoolLayer((2, 2), stride=2, method='average'),
    FlatLayer()
]

LeNet_dense_layers = [Layer(256), Layer(120, 'relu'), Layer(84, 'relu'), Layer(10)]
LeNet_fully_connected = MultiLayerPerceptron(LeNet_dense_layers)

LeNet_5 = ConvNeuralNet(LeNet_layers, LeNet_fully_connected)
LeNet_5.he_initialization()
LeNet_5.forward(reshaped_images[0])
```

```python
# Set up the subplots
fig, axes = plt.subplots(2, 3, figsize=(12, 8))  # 2 rows, 3 columns

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot each array in a separate subplot
for i, ax in enumerate(axes):
    ax.imshow(LeNet_5.layers[0].output[i], cmap='gray')
    ax.set_title(f'Feature Map {i + 1} / {len(LeNet_5.layers[0].output)}')

# Adjust layout to prevent clipping of titles
plt.tight_layout()

# Show the plot
plt.show()
```
